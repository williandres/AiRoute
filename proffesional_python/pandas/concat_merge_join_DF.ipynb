```python
import pandas as pd
import numpy as np
```

# Concat

```python
df1 = pd.DataFrame({
    'A': ['A0','A1','A2','A3'],
    'B': ['B0','B1','B2','B3'],
    'C': ['C0','C1','C2','C3'],
    'D': ['D0','D1','D2','D3']
    })

df2 = pd.DataFrame({
    'A': ['A4','A5','A6','A7'],
    'B': ['B4','B5','B6','B7'],
    'C': ['C4','C5','C6','C7'],
    'D': ['D4','D5','D6','D7']
    })
```

```python
pd.concat([df1,df2]) # bad indexing
```

```python
pd.concat([df1,df2], ignore_index = True)
```

```python
pd.concat([df1,df2], axis = 1)
```

# Merge


```python
left = pd.DataFrame(
{
    'key' : ['K0','K1','K2','K3'],
    'A': ['A0','A1','A2','A3'],
    'B': ['B0','B1','B2','B3']
})

right = pd.DataFrame(
{
    'key' : ['K0','K1','K2','K3'],
    'C': ['C0','C1','C2','C3'],
    'D': ['D0','D1','D2','D3']
})
```

```python
left.merge(right, on = 'key')
```

```python
left = pd.DataFrame(
{
    'key_1' : ['K0','K1','K2','K3'],
    'A': ['A0','A1','A2','A3'],
    'B': ['B0','B1','B2','B3']
})

right = pd.DataFrame(
{
    'key_2' : ['K0','K1','K2','K3'],
    'C': ['C0','C1','C2','C3'],
    'D': ['D0','D1','D2','D3']
})
```

```python
left.merge(right, on = 'key') #The keys are different
```

```python
left.merge(right, left_on = 'key_1', right_on = 'key_2')  #fixed
```

```python
left = pd.DataFrame(
{
    'key_1' : ['K0','K1','K2', np.nan],
    'A': ['A0','A1','A2','A3'],
    'B': ['B0','B1','B2','B3']
})

right = pd.DataFrame(
{
    'key_2' : ['K0','K1','K2','K3'],
    'C': ['C0','C1','C2','C3'],
    'D': ['D0','D1','D2','D3']
})
```

```python
left.merge(right, left_on = 'key_1', right_on = 'key_2')  #Â¿nan?
```

```python
left.merge(right, left_on = 'key_1', right_on = 'key_2', how = 'right')
```

# Join

```python
left = pd.DataFrame({
    'A': ['A0','A1','A2'],
    'B': ['B0','B1','B2']},
    index = ['k0','k1','k2']
    )

right = pd.DataFrame({
    'C': ['C0','C1','C2'],
    'D': ['D0','D1','D2']},
    index = ['k0','k2','k3']
    )
```

```python
#LEFT JOIN
left.join(right)
```

```python
#INNER JOIN
left.join(right, how = 'inner')
```

```python
#OUTER JOIN
left.join(right, how = 'outer')
```

```python
#RIGHT JOIN
left.join(right, how = 'right')
```
