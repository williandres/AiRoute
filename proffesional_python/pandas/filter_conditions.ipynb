```python
import pandas as pd
import numpy as np
```

```python
mr = pd.read_html('https://en.wikipedia.org/wiki/List_of_cities_by_homicide_rate')
top_homicide = mr[0].copy()
top_homicide.iloc[:10]
```

```python
colombia = top_homicide['Country'] == 'Colombia'
```

```python
top_homicide[colombia]
```

```python
top_homicide[top_homicide['Homicides (2019)'] > 1000]
```

```python
homicides = top_homicide['Homicides (2019)'] > 1000
top_homicide[homicides & colombia]
```

```python
top_homicide[~homicides & colombia]
```

```python
top_homicide[~homicides & ~colombia]
```

```python
top_homicide[top_homicide['City'].str.contains('ca')]
```

```python
top_homicide[(top_homicide['Homicides (2019)'] > 1000) & (top_homicide['Population (2019)'] > 1_000_000)]
```
