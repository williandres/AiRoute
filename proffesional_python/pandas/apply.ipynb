```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('sources/all_games.csv')
df_games = df.copy()
df_games['user_review'][df_games['user_review'] == 'tbd'] = np.nan
df_games['user_review'] = df_games['user_review'].astype('float32')
df_games.drop_duplicates(subset = ['name'], keep = 'first', inplace = True)
df_games.head()
```

```python
def two_times(value):
    return value * 2
```

```python
df_games['2x']= df_games['meta_score'].apply(two_times)
df_games.head()
```

```python
df_games['3x']= df_games['meta_score'].apply(lambda x: x * 3)
df_games.head()
```

```python
df_games['platform'] = df_games['platform'].astype('string')
```

```python
df_games.info()
```

```python
df_games['rating'] = df_games.apply(lambda x: x['meta_score'] * 2 if x['name'] == 'SoulCalibur' else 0, axis = 1)
```
