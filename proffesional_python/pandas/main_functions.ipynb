```python
df = pd.read_csv('sources/all_games.csv')
df_games = df.copy()
```

```python
df_games.info()
```

```python
df_games['user_review'][df_games['user_review'] == 'tbd'] = np.nan
```

```python
df_games.dropna(inplace = True)
```

```python
df_games['user_review'] = df_games['user_review'].astype('float32')
```

```python
top_games_by_users = df_games.sort_values('user_review', ascending = False) 
top_games_by_users.head(15)
```

```python
top_games_by_users['user_review'].skew() #El rating segun los usuarios tiende a sér negativo
```

```python
top_games_by_users['meta_score'].skew() #El rating segun los medios tiende a sér un poco menos negativo
```
