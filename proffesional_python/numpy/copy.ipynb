```python
import numpy as np
```

```python
arr = np.arange(0,20,2)
arr
```

```python
piece = arr[0:5]
piece
```

```python
piece[:] = 0 # Cambiar todos los valores a 0
piece
```

```python
arr #xddd - Se modifica la estructura general aunque la intencion sea cambiar otra referencia
```

```python
arr_copy = arr.copy() # Evitamos que al moficar algo se modifique el objeto padre
arr_copy[:] = 100
arr_copy
```

```python
arr # No se modifico, se usaron otros espacios de memoria
```

```python
# Espacios de memoria, que hace referencia a que. Muy util para otros problemas.
print("Original array: ",id(arr))
print("Copy array: ",id(arr_copy))
```
