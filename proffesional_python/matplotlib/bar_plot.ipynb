```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
country = ['INDIA','JAPAN','MEXICO','COLOMBIA','GERMANY']
population = [1000,800,900,1000,300]
```

```python
plt.bar(country,population,width=0.5,color=['red','blue','yellow'])#(x,y, args*)
plt.xticks(np.arange(5),('India','Japon','Mexico','Colombia','Alemania'),rotation=45)#COnfiguraciones del eje x
plt.show()
```

```python
plt.barh(country,population)
plt.show()
```

```python
fig, ax= plt.subplots(figsize=(5,5))
plt.style.use('_mpl-gallery')
plt.bar(country, population, width= 0.5, color= ['#FA07FF', '#07DAFF', '#FFB207', '#07FF80','#9207FF'])
plt.xticks(np.arange(5),('India', 'Japon', 'Mexico','Colombia','Alemania'), rotation= 35, fontsize= 13)#Configutaciones en el eje x
plt.xlabel('Country',fontsize= 15)
plt.ylabel('Population', fontsize= 14)
plt.title('Population by country', fontsize= 15)
```
