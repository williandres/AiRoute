```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
plt.subplot(1,2,1)# Rows - Columns - Index
plt.plot(x,y)
plt.subplot(1,2,2)# Rows - Columns - Index
plt.hist(y)
plt.show()
```

```python
plt.subplot(1,2,1)# Rows - Columns - Index
plt.plot(x,y,'r--')
plt.plot(y,x,'b:')
plt.subplot(1,2,2)# Rows - Columns - Index
plt.pie(y)
plt.show()
```

```python
plt.subplot(2,1,1)# Rows - Columns - Index
plt.plot(x,y,'r--')
plt.plot(y,x,'b:')
plt.subplot(2,1,2)# Rows - Columns - Index
plt.pie(y)
plt.show()
```
