```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

# Metodo orientado a objetos

Sobre un lienzo o figura se van a indtroducir graficas/objetos/axis. Esto permite que cada objeto sea personalziable.

```python
fig = plt.figure() # "Lienzo"  ----> axes = graficas
axes = fig.add_axes([0.1, 0.1, 0.2, 0.9])# add_axes([Posicion horizontal en la figura, Posicion vertical en la figura, Tamaño horizontal, Tamaño vertical])
axes.plot(x,y,'b')
fig.show()
```

```python
fig = plt.figure() 
axes = fig.add_axes([0.1, 0.1, 0.8, 0.9])
axes2 = fig.add_axes([0.1, 0.1, 0.4, 0.3])

axes.plot(x,y,'b')
axes2.plot(y,x,'r')

fig.show()
```

```python
fig = plt.figure() 
axes = fig.add_axes([0.1, 0.1, 0.8, 0.9])
axes2 = fig.add_axes([0.3, 0.5, 0.4, 0.3])

axes.plot(x,y,'b')
axes2.plot(y,x,'r')
axes2.set_facecolor('gray')

fig.show()
```
