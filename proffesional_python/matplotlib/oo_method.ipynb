```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
fig = plt.figure() # "Lienzo"  ----> axes = graficas
axes = fig.add_axes([0.1, 0.1, 0.5, 0.9])# ass_axes([Posicion de la figura, Tamaño vertical, Tamaño horizontal, ])
axes.plot(x,y,'b')
fig.show()
```
