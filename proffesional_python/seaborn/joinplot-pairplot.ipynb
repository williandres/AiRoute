```python
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
tips = sns.load_dataset("tips")
tips.head(2)
```

# JOIN-PLOT / PAIR-PLOT

Useful to quick analize for data entry


```python
#Grafica con mayor densidad de informaciom
sns.jointplot(data=tips, x='total_bill',y='tip',hue='sex',kind='hist',marginal_ticks=True,
                marginal_kws=dict(bins=25,fill=False,multiple='dodge'))
plt.show()
```

```python
#Relacion entre variables numericas automatica
sns.pairplot(data=tips,hue='sex',palette='coolwarm',corner=True,kind='scatter')
plt.show()
```
