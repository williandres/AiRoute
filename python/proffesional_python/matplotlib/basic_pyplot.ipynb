# Box plot / Diagrama de cajas
Distribucion de un conjunto de datos.

Un rectángulo, que representa el intervalo intercuartílico (el rango que va desde el primer cuartil hasta el tercero).
Dos “bigotes” que salen del rectángulo y que representan los datos extremos (los valores más altos y más bajos que están fuera del intervalo intercuartílico).
Una línea horizontal dentro del rectángulo que representa la mediana (el valor que divide al conjunto de datos en dos partes iguales).

Los cuartiles son puntos de corte que dividen a un conjunto de datos en cuatro partes iguales. Los cuartiles se utilizan a menudo en el gráfico de caja para describir la distribución de un conjunto de datos.

Los cuartiles se denominan Q1, Q2 y Q3. Q1 es el primer cuartil, Q2 es el segundo cuartil (también conocido como mediana) y Q3 es el tercer cuartil.

Para calcular los cuartiles de un conjunto de datos, primero debes ordenar los datos de menor a mayor. Luego, puedes calcular Q1, Q2 y Q3 siguiendo estos pasos:

Q1: el primer cuartil es el valor que divide al conjunto de datos en dos partes iguales, con el 25% de los datos por debajo de él y el 75% de los datos por encima de él.
Q2: el segundo cuartil es la mediana del conjunto de datos, es decir, el valor que divide al conjunto de datos en dos partes iguales, con el 50% de los datos por debajo de él y el 50% de los datos por encima de él.
Q3: el tercer cuartil es el valor que divide al conjunto de datos en dos partes iguales, con el 75% de los datos por debajo de él y el 25% de los datos por encima de él.
En el gráfico de caja, el rectángulo del gráfico representa el intervalo intercuartílico, es decir, el rango que va desde Q1 hasta Q3. Los bigotes del gráfico representan los datos extremos (los valores más altos y más bajos que están fuera del intervalo intercuartílico). La línea horizontal dentro del rectángulo representa la mediana (Q2).

```python
plt.boxplot(x)
plt.show
```
