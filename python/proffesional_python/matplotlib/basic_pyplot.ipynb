# Box plot / Diagrama de cajas
Distribucion de un conjunto de datos.

Un rectngulo, que representa el intervalo intercuartlico (el rango que va desde el primer cuartil hasta el tercero).
Dos bigotes que salen del rectngulo y que representan los datos extremos (los valores ms altos y ms bajos que estn fuera del intervalo intercuartlico).
Una lnea horizontal dentro del rectngulo que representa la mediana (el valor que divide al conjunto de datos en dos partes iguales).

Los cuartiles son puntos de corte que dividen a un conjunto de datos en cuatro partes iguales. Los cuartiles se utilizan a menudo en el grfico de caja para describir la distribucin de un conjunto de datos.

Los cuartiles se denominan Q1, Q2 y Q3. Q1 es el primer cuartil, Q2 es el segundo cuartil (tambin conocido como mediana) y Q3 es el tercer cuartil.

Para calcular los cuartiles de un conjunto de datos, primero debes ordenar los datos de menor a mayor. Luego, puedes calcular Q1, Q2 y Q3 siguiendo estos pasos:

Q1: el primer cuartil es el valor que divide al conjunto de datos en dos partes iguales, con el 25% de los datos por debajo de l y el 75% de los datos por encima de l.
Q2: el segundo cuartil es la mediana del conjunto de datos, es decir, el valor que divide al conjunto de datos en dos partes iguales, con el 50% de los datos por debajo de l y el 50% de los datos por encima de l.
Q3: el tercer cuartil es el valor que divide al conjunto de datos en dos partes iguales, con el 75% de los datos por debajo de l y el 25% de los datos por encima de l.
En el grfico de caja, el rectngulo del grfico representa el intervalo intercuartlico, es decir, el rango que va desde Q1 hasta Q3. Los bigotes del grfico representan los datos extremos (los valores ms altos y ms bajos que estn fuera del intervalo intercuartlico). La lnea horizontal dentro del rectngulo representa la mediana (Q2).

```python
plt.boxplot(x)
plt.show
```
