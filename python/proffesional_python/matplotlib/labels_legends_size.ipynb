```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
# Oriented object method
fig, axes = plt.subplots(1,2,figsize=(6,6)) # figsize(horizontal, vertical)
axes[0].plot(x,y,label="$sin(x)$")# -> Notacion matemática
axes[0].set_title('Relacion X - Y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].legend()
axes[1].plot(y,x)
axes[1].set_title('Relacion Y - X')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X')
```
