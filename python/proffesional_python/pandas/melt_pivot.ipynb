# Pivot
Transforma los valores de determinadas columnas o filas en los índices de un nuevo DataFrame, y la intersección de estos es el valor resultante.

```python
df_books = pd.read_csv('sources/bestsellers-with-categories.csv',sep=',',header=0)
```

```python
df_books.head()
```

```python
df_books.pivot_table(index='Author',columns='Genre',values='User Rating')
```

```python
df_books.pivot_table(index='Genre',columns='Year', values='User Rating' ,aggfunc='max')
```

# Melt 
Toma las columnas del DataFrame y las pasa a filas, con dos nuevas columnas para especificar la antigua columna y el valor que traía.

```python
df_books[['Name','Genre']].head(5)
```

```python
df_books.melt(id_vars='Year',value_vars='Genre')
```
