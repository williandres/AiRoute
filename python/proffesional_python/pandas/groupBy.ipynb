```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('sources/all_games.csv')
df_games = df.copy()
df_games['user_review'][df_games['user_review'] == 'tbd'] = np.nan
df_games['user_review'] = df_games['user_review'].astype('float32')
df_games.drop_duplicates(subset = ['name'], keep = 'first', inplace = True)
df_games
```

```python
df_games.groupby('platform').count()
```

```python
df_games.groupby('platform').mean()
```

```python
df_games.groupby('platform').min()
```

```python
df_games.groupby('platform').max().sort_values('meta_score')
```

```python
df_games.groupby('platform').max().sort_values('meta_score').reset_index()
```

```python
df_games.groupby('platform').agg(['min','max'])
```

```python
df_games.groupby('platform').agg({'meta_score': ['min','max'], 'user_review': 'sum'})
```
