```python
import numpy as np
import random as rm
```

```python
ra = lambda n: [rm.randint(0,10) for x in range(n)]
```

# ARRAY DIMENSIONS

```python
scalar = np.array(54)
scalar.ndim
```

```python
vector = np.array([1,2,3])
vector.ndim
```

```python
matrix = np.array([ra(3), ra(3), ra(3)])
print(matrix, "\n")
print("1: ", matrix[:,1])
print("2: ", matrix[2], "\n")
print(f'Dimensions: {matrix.ndim}')
```

```python
tensor = np.array([[ra(3),ra(3),ra(3)],[ra(3),ra(3),ra(3)],[ra(3),ra(3),ra(3)]])
print(tensor, "\n")
print(f'Dimensions: {tensor.ndim}')
```

# Add or delete array dimensions

```python
vector = np.array([1,2,3], ndmin = 10)
print(vector, "\n")
print("Dimensions: ", vector.ndim)
```

```python
# Expand in axis
expand = np.expand_dims(np.array([1,2,3]), axis = 0)
print(expand, "\n")
print("Dimensions: ", expand.ndim)
```

```python
print(vector, "\n")
print("Dimensions: ", vector.ndim)

print('\n',"-----"*5, '\n')

vector_2 = np.squeeze(vector) #Correct dimension
print(vector_2, "\n")
print("Dimensions: ", vector_2.ndim)
```
