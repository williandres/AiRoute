```python
import numpy as np
```

```python
arr = np.linspace(1,20,5, dtype = 'int8')
print(arr)
```

```python
indexes = arr > 5
indexes
```

```python
arr[indexes]
```

```python
arr[indexes & (arr < 20)]
```

```python
arr_copy  = arr.copy()
arr_copy[(arr_copy > 5) & (arr_copy < 20)] = 0
```

```python
print(arr, id(arr))
print(arr_copy, id(arr_copy))
```

```python
arr = np.arange(1,11)

def number_is_even(num):
    return num % 2 == 0


number_is_even(arr)

lambda_func = lambda x: x % 2 == 0

lambda_func(arr)
```
