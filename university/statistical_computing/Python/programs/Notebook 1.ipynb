### Ejemplos:

```
#1 - Primera fila contiene los nombres de las variables.
plantulas1 = pd.read_excel(datasets + "sample.xlsx",sheet_name=12,header=0)
plantulas1.info()
plantulas1.head(10)
```

```
#2
plantulas2 = pd.read_excel(datasets + "sample.xlsx",sheet_name="Plantulas",header=0)
plantulas2.info()
plantulas2.tail(10)
```

```
#3 - Columnas "A", "B", "C" y "E"
plantulas3 = pd.read_excel(datasets + "sample.xlsx",sheet_name="Plantulas",usecols="A:C,E",header=0)
plantulas3.info()
plantulas3.head(10)
```

```
#4 - Primera 35 filas
plantulas4 = pd.read_excel(datasets + "sample.xlsx",sheet_name="Plantulas",nrows=35,header=0)
plantulas4.info()
plantulas4.head(10)
```

```
#5 - Nombres personalizados a las variables
plantulas5 = pd.read_excel(datasets + "sample.xlsx",sheet_name="Plantulas",header=0,
                           names=["IdParcela","N_I","IdPlantula","N_i","Altura"])
plantulas5.info()
plantulas5.head(10)
```

```
#6 - Formato text, numeric, text, numeric y numeric, respectivamente.
plantulas6 = pd.read_excel(datasets + "sample.xlsx",sheet_name="Plantulas",header=0,
                          names=["IdParcela","N_I","IdPlantula","N_i","Altura"],
                          dtype={"IdParcela":str,"N_I":np.int64,"IdPlantula":str,
                                 "N_i":np.float64,"Altura":np.float64})
plantulas6.info()
plantulas6.head(10)
```

```
#7 - La segunda variable se excluye y el tipo de la cuarta variable se especifica a partir de una inspección de los datos.
plantulas7 = pd.read_excel(datasets + "sample.xlsx",sheet_name="Plantulas",header=0,usecols=[0,2,3,4],
                           names=["IdParcela","IdPlantula","N_i","Altura"],
                           dtype={"IdParcela":str,"IdPlantula":str,"Altura":np.float64})
plantulas7.info()
plantulas7.head(10)
```

```
#8 -  Los datos están separados por " "
stocks1 = pd.read_table(datasets + "stocks.dlm",delimiter=" ",header=0)
stocks1.info()
stocks1.head(10)
```

```
#9 -  Separador decimal es "," y el separador de agrupamiento es ".".
stocks2 = pd.read_table(datasets + "stocks.dlm",delimiter=" ",decimal=",",thousands=".",header=0)
stocks2.info()
stocks2.head(10)
```

```
#10 - Especificando el formato en el que se deben leer las variables.
from datetime import datetime
dateparse = lambda x: datetime.strptime(x,"%d%b%y")

stocks3 = pd.read_table(datasets + "stocks.dlm",delimiter=" ",decimal=",",thousands=".",header=0,
                        dtype={"A":str,"B":str,"C":np.float64,"D":np.float64,"E":np.float64,
                              "F":np.float64,"G":np.float64,"H":np.float64},
                        parse_dates=["B"], date_parser=dateparse)

stocks3.info()
stocks3.head(10)
```
