### Cuantas mujeres tienen información completa (es decir, 0 valores perdidos) sobre

### Cuantas mujeres tienen información sobre tres de las cuatro ocasiones en las que
se midió la presencia o ausencia de amenorrea?

```
acetato_100 <- subset(data, Dose == 0)
zas <- sqldf("SELECT acetato_100.*, count(*) as count
              FROM acetato_100
              WHERE amenhorrea is null
              GROUP BY ID")

tyr <- subset(zas, count == 1)
str(tyr)
#48  mujeres solo tienen informacion en 3 ocasiones
```

### Cuantas mujeres tienen información sobre dos de las cuatro ocasiones en las que

```
tyr <- subset(zas, count == 2)
str(tyr)
#68  mujeres solo tienen informacion en 2 ocasiones
```

### Cuantas mujeres tienen información sobre una de las cuatro ocasiones en las que

### Cuantas mujeres siempre presentaron amenorrea?

```
acetato_150_not_null <- subset(data, Dose == 1 & !(is.na(amenhorrea)))

acetato <- sqldf("SELECT *, sum(amenhorrea) as sum, count(*) as count, count(*) - sum(amenhorrea) as nunca
                  FROM acetato_150_not_null
                  GROUP BY ID")

resp <- subset(acetato, nunca == 0 & count == 4 )
str(resp)
resp
#28 mujeres siempre tuvieron amenhorrea en las 4 ocasiones
```
