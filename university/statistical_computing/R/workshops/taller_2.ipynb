### ¿Cuántas películas tienen solo 1 director(a)? ¿Cuántas películas tienen 2 directores(as)? ¿Cuántas películas tienen 3 o más directores(as)?

### Cree un conjunto de datos con las películas que tienen solo 1 director(a)

Use la variable Director para crear una variable llamada Director2 teniendo en cuenta los siguientes parámetros: (a) Deben aparecer los nombre(s) y luego los apellido(s) del(de la) director(a) de la película separados por espacios sencillos y sin espacios al inicio ni al final; (b) Las iniciales de cada nombre y de cada apellido deben estar en letras mayúsculas.

```
#Subconjunto de movies
director_name <- subset(movies, n_directors == 1, select = -c(n_directors))
head(director_name)
```

```
director_name <- within(director_name,{
        base <- str_split(director_name$Director, ",|\\.")
        Director2 <- str_c(sapply(base, tail, 1),sapply(base, head, 1))
        Director2 <- str_squish(Director2)
        Director2 <- str_to_title(Director2)
        rm(base)
})
head(director_name, n = 3)
tail(director_name, n = 3)
```

### Filtros

```
#Filtro Directores(as)
busca_director <- function(nombre_director) return (subset(director_name, Director2 == nombre_director))
```

```
#Limpiar a movies
movies <- within(movies,{
    Pais <- str_squish(Pais)
    rm(n_directors)
})
head(movies)
```

```
#Filtro Paises
busca_pais <- function(nombre_pais) return (subset(movies, Pais == nombre_pais))
```

### ¿Qué porcentaje de la cartera (número de créditos) corresponde a créditos “malos”?

Los créditos considerados "buenos" y "malos" se identifican como class = 0 y class = 1

```
#El 29.14% de numero de creditos de la cartera son malos
mytable(Number_of_existing_credits_at_this_bank	 ~ Class,data=credit,ord="freq")
```

El 29.14% de número de créditos de la cartera son malos

### ¿Qué porcentaje de la cartera (valor de los créditos) corresponde a créditos "malos"?

```
#El 36.12% del valor de los creditos de la cartera son malos
mytable(Credit_amount ~ Class,data=credit,ord="freq")
```

El 36.12% del valor de los creditos de la cartera son malos

### El porcentaje de la cartera (número de créditos) que corresponde a créditos "malos" es menor para los clientes con "buen" historial crediticio?

Historial de crédito
A30 : no se tomaron créditos/ todos los créditos se devolvieron debidamente
A31 : todos los créditos en este banco reembolsados debidamente
A32: créditos existentes reembolsados debidamente hasta ahora
A33 : retraso en el pago en el pasado
A34 : cuenta crítica/ otros créditos existentes (no en este banco)

```
#Historial de crédito
#A30 : no se tomaron créditos/ todos los créditos se devolvieron debidamente
#A31 : todos los créditos en este banco reembolsados debidamente
#A32: créditos existentes reembolsados debidamente hasta ahora
#A33 : retraso en el pago en el pasado
#A34 : cuenta crítica/ otros créditos existentes (no en este banco)

#Si, las mejores historias crediticias ocupan un menor porcentaje de numero de créditos malos (A30 , A31)  -> 20%. 
#Con la excepción de A32 que a pesar de no estar tan mal ocupa un 50%
#El 30% restante lo ocupan los peores historiales créditicios (A34, A33).

mytable(Number_of_existing_credits_at_this_bank	 ~ Credit_history,data=credit,subset= c(Class==1),ord="freq")
```

SI, las mejores historias crediticias ocupan un menor porcentaje de créditos malos (A30 , A31)  -> 20%. 

Con la excepción de A32 que a pesar de no ser una mala categoria ocupa casi un 50%

El 30% restante lo ocupan los peores historiales créditicios (A34, A33).

### El porcentaje de la cartera (en valor) que corresponde a créditos "malos" es menor para los clientes con "buen" historial crediticio?

```
#Sí, las mejores historias crediticias ocupan un menor porcentaje de valor de créditos malos (A30 , A31)  -> 20%. 
#Con la excepción de A32 que a pesar de no ser una mala categoría, ocupa más de un 50%
#Las peores categorías (A33, A34) ocupan poco más del 25%.
#Así que se podría determinar que los menores valores de créditos se asocian a buenas historias crediticias, sin embargo, la diferencia no se acentúa tanto como en el número de créditos.

mytable(Credit_amount ~ Credit_history,data=credit,subset= c(Class==1),ord="freq")
```

Sí, las mejores historias crediticias ocupan un menor porcentaje de valor de créditos malos (A30 , A31)  -> 20%. 
Con la excepción de A32 que a pesar de no ser una mala categoría, ocupa más de un 50%
Las peores categorías (A33, A34) ocupan poco más del 25%.
Así que se podría determinar que los menores valores de créditos se asocian a buenas historias crediticias, sin embargo, la diferencia no se acentúa tanto como en el número de créditos.
