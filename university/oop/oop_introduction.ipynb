Abstracción: La abstracción es la capacidad de enfocarse en los aspectos importantes de un objeto y ignorar los detalles irrelevantes. En POO, una clase es una abstracción que define las características y comportamientos de un conjunto de objetos. La abstracción permite modelar conceptos del mundo real de una manera que sea fácil de entender y manipular en un programa.

Encapsulación: La encapsulación es el proceso de ocultar la complejidad de un objeto y proteger su estado interno. En POO, una clase encapsula sus datos y métodos para que no sean accesibles desde fuera de la clase. Esto reduce la complejidad y hace que el objeto sea más fácil de usar y mantener.

Herencia: La herencia es la capacidad de crear nuevas clases a partir de clases existentes. La clase existente se convierte en la clase base o padre, y la nueva clase se convierte en la clase derivada o hija. La clase hija hereda los atributos y métodos de la clase padre y puede agregar sus propios atributos y métodos. La herencia permite la reutilización de código y la creación de jerarquías de clases que reflejen las relaciones del mundo real.

Polimorfismo: El polimorfismo es la capacidad de un objeto para tomar varias formas. En POO, el polimorfismo se logra mediante la implementación de métodos con el mismo nombre en diferentes clases. Los objetos de estas clases pueden responder de manera diferente a un mismo método, lo que permite que un programa sea más flexible y extensible.

### Clase abstracta

Es una clase que no puede ser instanciada, es decir, no se pueden crear objetos directamente a partir de ella. En su lugar, una clase abstracta proporciona una plantilla o un molde para crear subclases que sí se pueden instanciar. Las clases abstractas suelen contener métodos abstractos, que son métodos que no tienen una implementación definida en la clase abstracta, sino que deben ser implementados en las subclases que heredan de ella. Las clases abstractas se utilizan para definir una interfaz común entre un conjunto de subclases relacionadas, lo que ayuda a simplificar y organizar el código.

Solo pueden ser heredas, no contienen implementación, no se instancian. Solo contienen métodos abstractos.

Las clases abstractas establecen un contrato de diseño sobre las subclases que heredan, es decir, establece un criterio de diseño para las clases heredadas.

Cuando este obliga a sus subclases a sobrescribir los método abstractos. Podemos hacer la analogía tomando como método el área de una figura, este método abstracto cambia para un circulo como para un cuadrado (deben implementar el método de forma especifica).

### Decoradores de Python: @property, @classmethod y @staticmethod

@property: En Python, el decorador @property se utiliza para convertir un método en un atributo de solo lectura. Esto significa que se puede acceder al método como si fuera un atributo de una clase, sin necesidad de agregar paréntesis al final del nombre del método. Esto es útil para ocultar la complejidad del método detrás de un nombre de atributo simple y fácil de entender.

@classmethod: El decorador @classmethod se utiliza en Python para definir un método que puede ser llamado en una clase y no en una instancia de la clase. A diferencia de un método de instancia, un método de clase tiene acceso a la clase misma y no a las instancias de la clase. Esto puede ser útil para crear métodos de fábrica que crean nuevas instancias de una clase y para realizar operaciones en la clase en sí.

@staticmethod: El decorador @staticmethod se utiliza en Python para definir un método que no tiene acceso a la clase o a las instancias de la clase. Un método estático se define como una función dentro de una clase, pero no tiene relación con la instancia de la clase o sus atributos. Esto es útil cuando se necesita una función en una clase, pero la función no necesita interactuar con los atributos de la clase o las instancias de la clase.
