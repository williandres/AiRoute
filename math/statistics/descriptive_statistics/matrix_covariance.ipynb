```
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler

iris = sns.load_dataset('iris')
```

```
sns.pairplot(iris, hue = 'species')
```

```
scaler = StandardScaler() #Escalar los datos, normalizar segun sus datos
scaled = scaler.fit_transform(
    iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
)
scaled.T # Traspuesta de datos para adaptarlo a las dimesiones requeridas para aplicar la matriz de covarianza
# Quitar el .T para probar el cambio
```

```
covariance_matrix = np.cov(scaled.T)# Aplicar la covarianza al arreglo de datos escalados (matriz de covarianza), metodo numpy
covariance_matrix
#https://numpy.org/doc/stable/reference/generated/numpy.cov.html
```

```
plt.figure(figsize=(10,10))
sns.set(font_scale=1.5)
hm = sns.heatmap(covariance_matrix,#matriz
                 cbar=True,#Escala de colores
                 annot=True,#Muestra los valores
                 square=True,#Matriz cuadrada
                 fmt='.2f',#formatos
                 annot_kws={'size': 12},#formatos
                 yticklabels=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'],
                 xticklabels=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
```
