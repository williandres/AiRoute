Dominio y rangos de la funcion expoencial:

La funcion: $$f\left(x\right)=a^x \:\:\:\:\:\ a\:>0\:,\:a\:\ne 1$$ 

tiene **dominio** $\mathbb{R}$ y **rango** (0, $\infty$). La recta y = 0 (el eje x) es una **asintona horizontal** de f. La grafica de f tiene una de las siguientes formas:


$$f\left(x\right)=a^x\:para\:\:0<a\:<1$$

```python
a = np.array([1/2,1/3,1/5,1/10]) #Se grafican cuatro funciones exponenciales con 0 < a < 1
x = np.linspace(-1,1,num=70)

def funciion_exp(a,x):
  for i in a:
    y = i**x
    ax.plot(x,y,label="Base: "+str(i))

fig, ax = plt.subplots()
fig.set_size_inches(7,7) #Ajusto la imagen de salida

funciion_exp(a,x)
ax.legend() #Muestra las etiquetas de cada funcion graficada
ax.grid() 
```

### Transformaciones de Funciones Exponenciales

Para mover una funcion exponecial para arriba o abajo, se le suma o resta respectivamente la unidad a mover:

**Ejemplo**: Mover dos unidades arriba la funcion $2^x$: $$g\left(x\right)=2+2^x$$ 

```python
x = np.linspace(-3,3,num=70)
def funcionExp(x):
  return 2**x
fig, ax = plt.subplots()
fig.set_size_inches(6,6)
y = funcionExp(x)
ax.plot(x,y,label="2**x")

y = 2+funcionExp(x)
ax.plot(x,y, label="2+2**x")

y = np.linspace(2,2,num=70)
ax.plot(x,y, label="2",linestyle= "--")

ax.legend()
ax.grid()
```

Para crear un reflejo en el eje x:
 
**Ejemplo**:  $-2^x$

```python
fig, ax = plt.subplots()
fig.set_size_inches(6,6)
y = funcionExp(x)
ax.plot(x,y,label="Original")
y = -funcionExp(x)
ax.plot(x,y, label="Reflejo en eje x")
ax.legend()
ax.grid()
```

Para desplazar a la izquierda o a la derecha,  se suma o se resta rerspectivamente en el exponente

**Ejemplo**: Mover dos unidades a la izquierda $2^x$:
$$g\left(x\right)=2^{x+2}$$

```python
fig, ax = plt.subplots()
fig.set_size_inches(6,6)
y = funcionExp(x)
ax.plot(x,y,label="Original")
y = funcionExp(x+2)
ax.plot(x,y, label="Se movio a la izquierda")
ax.legend()
ax.grid()
```

### Interes Compuesto (Ejemplo de funcion Exponencial)
Formula del interes compuesto:

$$A\left(t\right)=P\left(1+\frac{r}{n}\right)^{nt}$$
donde 
- $A\left(t\right)$ = cantidad despues de t años
- *P* = principal (Inversion inicial)
- *r* = tasa de interes por año
- *n* = numero de veces que el interes se capitaliza por año
- *t* = numero de años

**Calculo del rendimiento en porcentaje anual** (Si una inversion de capitaliza cada n meses, n dias, ect, ¿Cual es su equivalente real anual?)

$$r=\left(1+\frac{r}{n}\right)^{nt}-1$$

Explicacion de la deduccion: Precalculo: Matematicas para el calculo **James Stewart, Lothar Redlin, Saleem Watson** | Capitulo 4, seccion 4.1 - Funciones exponenciales


**Ejemplo:** Si tengo 1000 pesos a una tasa del 12% con una capitalizacion mensual, ¿Cuanto tendre en 3 años?
¿Cual es el rendimiento anual?

```python
def interesCompuesto(P,r,n,t):
  return P*((1+(r/n))**(n*t))

print(f'En 3 años tendria {round(interesCompuesto(1000,.12,12,3),2)} pesos')

def calculo_rendimiento_anual(r,n,t):
  return ((1+(r/n))**(n*t))-1

print(f'El rendimiento equivalente anual seria del {round(calculo_rendimiento_anual(0.12,12,1)*100,2)}% ')
```

##Funciones Logaritmicas
Toda función exponencial $f\left(x\right)=a^x$ , con a > 0 y a $\ne $ 1, es una función uno a uno por la prueba de la Recta Horizontal y por tanto tiene una función inversa. La función inversa $f^{-1}$ se denomina **función logarítmica con base a** y se denota con
$\log _a$ . Recuerde de la Sección 2.6 que $f^{-1}$ está deﬁnida por

$$f^{-1}\left(x\right)=y\:\:\:\:\:asi\:que\:\:\:\:f\left(y\right)=x$$

Esto lleva a la siguiente deﬁ nición de la función logarítmica.



> Sea a un número positivo con $a\:\ne 1$. La función logarítmica con base a,
denotada por $\log _a$ , está definida por $$\log _a\left(x\right)=y\:\:\:\:asi\:que\:\:\:\:a^y=x$$ 
Por lo tanto, $\log _a\left(x\right)$ es el exponente al cual la base *a* debe ser elevado para obtener *x*.

>Captura de pantalla de 2022-02-02 10-49-39.png

```python
x = np.linspace(0.001,10, num=100)
y = np.log10(x)
fig, ax = plt.subplots()
fig.set_size_inches(5,5)
ax.plot(x,y,label="logarimo")
ax.plot(y,x,label="Exponecial") #Se puede usar los mismo datos que el logaritmo pero invertidos (ya que el exponente es la funcion inversa del logaritmo)
x = np.linspace(y[0],x[-1], num=100)
ax.plot(x,x,linestyle= "--",label="Recta x = y")
ax.legend()
ax.grid()
```

####La Ley del Olvido
En especial se pueden usar las expresiones logaritmicas se pueden utilizar para modelar comportamientos del ser humano: como por ejemplo **Hermann Ebbinghaus** (1850-1909) modelo de acuerdo a los estudio que el realizo una formula sobre fenomeno de la memoria y cuanto tendemos a olvidar con el tiempo:

Modelo la siguiente ecuacion: $$log\left(P\right)=log\left(P_0\right)-c\:log\left(t+1\right)$$
Donde: 

- $P_0$ es el nivel aprendido inicial
- $t$ tiempo transcurrido (en meses)
- $c$ es una constante que depende del tipo de tarea
- $P$ es el nivel de rendimento pasado $t$ tiempo

Despejando se obtiene la siguiente formula:
$$P=\frac{P_0}{\left(t+1\right)^c}$$

**Ejemplo**: si tengo un 90 en una examen, cuanto sacaria en dos meses y un año? En una actividad con  $c = 0.2$

```python
def ley_olvido(Po,t,c):
  return Po/((t+1)**c)
print(f'En dos meses sacarias: {ley_olvido(90,2,0.2)}')
print(f'En un año sacarias: {ley_olvido(90,12,0.2)}')
meses = 4*12
x = np.linspace(0,meses, num=meses)
Po = np.linspace(90,90,num=meses)
c = np.linspace(0.2,0.2,num=meses)
fig, ax = plt.subplots()
fig.set_size_inches(5,5)
y = ley_olvido(Po,x,c)
ax.plot(x,y,label="Grafica del olvido")
ax.plot(np.linspace(2,2,num=meses),y,label="2 meses",linestyle= "--")
ax.plot(np.linspace(12,12,num=meses),y,label="1 año",linestyle= "--")
ax.legend()
ax.grid()
```
