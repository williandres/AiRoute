```python
from sympy import *
import matplotlib.pyplot as plt
import numpy as np
```

$$f\left(x\right)=x^4 - 3x^2$$

```python
f = lambda x: x**4 - 3*x**2

x = np.linspace(-2,2,100)
plt.plot(x,f(x),"r")
```

```python
init_printing( use_latex='mathjax' )
x = Symbol('x')
expr = x**4 - 3*x**2
# ^^ esta es nuestra expresion
expr_d = Derivative(expr, x, evaluate=True)
# ^^ esta es la derivada de nuestra expresion
```

```python
Eq(expr_d, 0)
```

# Puntos criticos

Ceros de la funcion derivada

```python
solve(Eq(expr_d, 0))
```
