# Implementaci√≥n de la derivada

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
  return np.sin(x)

x = np.linspace(-10,10, num=1000)

def df(f):
  h=0.000001
  return (f(x+h)-f(x))/h

y = f(x)


plt.plot(x,y, 'b')
plt.plot(x,df(f), 'r')
plt.grid()
```

```python
def f(x):
  return x**2

x = np.linspace(-10,10, num=1000)

def df(f):
  h=0.000001
  return (f(x+h)-f(x))/h

y = f(x)


plt.plot(x,y, 'b')
plt.plot(x,df(f), 'r')
plt.grid()
```
