# Dependencias

```python
import numpy as np
import matplotlib.pyplot as plt

N = 1000
x = np.linspace(-5,5, num=N)
```

# Derivada discreta

```python
def df(f):
  h=0.000001
  return (f(x+h)-f(x))/h
```

## Funci√≥n lineal
$$f(x)=mx+b$$

Derivada

$$f'(x)=m$$

Para redes neuronales.

Descenso del gradiente. Back propagation.


Regresion logistica y lineal.

```python
def f(x):
  return x

plt.plot(x, f(x), 'b')
plt.plot(x,df(f), 'r')
plt.grid()
```

## Funci√≥n escal√≥n o de Heaviside
$$H(x) = 
     \begin{cases}
        0, &\quad \text{para, } x < 0 \\
        1,  &\quad\text{para. } x \ge 0 \\
     \end{cases}
$$

Derivada

$$f¬¥(x)=0$$

No mucho espacio para optimizar.

Categoricos. 

Modelos de redes neuronales sencillos.

Para Heaviside, existe un cambio en el punto x=0, donde pasa de valer 0 a 1, en ese momento si hay una ‚Äúpendiente‚Äù que es infinita, y hay una funci√≥n para representar su derivada que es la delta de dirac, una funci√≥n de distribuci√≥n de densidad
https://es.wikipedia.org/wiki/Delta_de_Dirac


```python
def H(x):
  Y = np.zeros(len(x))
  for idx,x in enumerate(x):
    if x>=0:
      Y[idx]=1
  return Y
    
N=1000


y = H(x)

plt.plot(x, H(x), 'b')
plt.plot(x,df(H), 'r')
plt.grid()
```

## Funci√≥n sigmoide

$$f(x)=\frac{1}{1-e^{-x}}$$

Derivada

$$f¬¥(x)=f(x)(1-f(x))$$

Clasificaciones de probabilidad.

Dificil de optimizar.

Satura el proceso del gradiente.

Banishing gradient (muerte de redes neuronales).

Graficar de manera discreta es demandante para el procesamiento.

```python
def f(x):
  return 1/(1 + np.exp(-x))
    

N=1000

y = f(x)

plt.plot(x, f(x), 'b')
plt.plot(x,df(f), 'r')
plt.grid()
```

## Funci√≥n tangente hiperb√≥lica

$$f(x)=\frac{2}{1+e^{-2x}}-1$$

Derivada

$$f¬¥(x)=1-f(x)^2$$

Procesos de ML.

Costo computacional alto.

```python
def f(x):
  return np.tanh(x)
    

N=1000

y = f(x)

plt.plot(x, f(x), 'b')
plt.plot(x,df(f), 'r')
plt.grid()
```

## Funci√≥n ReLU

$$R(x)=max(0,x)$$
 Tambi√©n se puede escribir como
$$R(x) = 
     \begin{cases}
        0, &\quad \text{para, } x < 0 \\
        x,  &\quad\text{para. } x \ge 0 \\
     \end{cases}
$$

Derivada

$$H(x) = 
     \begin{cases}
        0, &\quad \text{para, } x < 0 \\
        1,  &\quad\text{para. } x \ge 0 \\
     \end{cases}
$$

Para entrenamiento es muy util (en proceso de capas intermedias)

Es sensible a la muerte neuronal

```python
def f(x):
  return np.maximum(x,0)
    

N=1000

y = f(x)

plt.plot(x, f(x), 'b')
plt.plot(x,df(f), 'r')
plt.grid()
```
