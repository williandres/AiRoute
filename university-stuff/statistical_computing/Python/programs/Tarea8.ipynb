```python
##################################### PUNTO 2 ###########################################

principals = pd.read_table("https://datasets.imdbws.com/title.principals.tsv.gz", delimiter="\t",
                         nrows=5000, usecols=[0,2,3,4,5],
                         na_values=["", "NA", "\\N"],
                         dtype={"col1": str, "col2": str, "col3": str, "col4": str, "col5": str})


basics = pd.read_table("https://datasets.imdbws.com/title.basics.tsv.gz", delimiter="\t",
                         nrows=5000, na_values=["", "NA", "\\N"])


namess = pd.read_table("https://datasets.imdbws.com/name.basics.tsv.gz", delimiter="\t",
                         nrows=5000, usecols=[0,1,2,3,4],
                         na_values=["", "NA", "\\N"])
print(principals.info())
print(basics.info())
print(namess.info())
```
