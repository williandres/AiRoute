Los archivos season-1415.xlsx, season-1516.xlsx, season-1617.xlsx, season-1718.xlsx y season-1819.xlsx contienen información sobre los resultados de las temporadas 14/15, 15/16, 16/17, 17/18 y 18/19, respectivamente, de la primera división del futbol inglés. ### Data sets en el directorio

```
 dataset <- list.files(path="/work/AiRoute/university-stuff/statistical_computing/sources/t1", pattern=NULL, all.files=FALSE,
    full.names=TRUE)

dataset
```

### Función que construye el dataframe

```
df <- function(route){
    seasonDf <- read_excel(
                    route,
                    skip = 1,
                    col_types = c("text", "guess","text","text","guess","guess","skip","skip","skip","skip","guess","guess","guess","guess","guess","guess","guess","guess","guess","guess","guess","guess","guess"),
                    col_names = c("Temporada","Dia","EquipioLocal","EquipoVisitante","FTHG","FTAG","ARBITRO","HS","AS","HST","AST","HF","AF","HC","AC","HY","AY","HR","AR")
                    )
    return(seasonDf)
}
```

### Ejemplo dataframe

```
head(df("/work/AiRoute/university-stuff/statistical_computing/sources/t1/season-1617.xlsx"))
```

### ¿Cuántos registros tiene cada conjunto?

```
 #Crea un diccionario donde cuenta los registros de cada archivo excel
 info_rows <- list()
 for (route in dataset){
    seasonDf <- df(route)                
    name <- substr(route, 65, 75)
    info_rows[[name]] <- nrow(seasonDf)
}
```

```
print("¿Cuántos registros tiene cada conjunto de datos?")
for (i in seq_along(info_rows)){
  print(info_rows[i])
}
```

## IMDb (Internet Movie Database)

Importe los primeros 5000 registros de los conjuntos de datos title.basics, title.principals y name.basics. Hágalo sin descargar ni descomprimir los archivos, y omitiendo las variables knownForTitles and ordering.

¿Cuántas variables tienen los conjuntos de datos? ¿De qué tipo son estas variables?

### Datasets

```
dfs <- c("https://datasets.imdbws.com/name.basics.tsv.gz","https://datasets.imdbws.com/title.basics.tsv.gz","https://datasets.imdbws.com/title.principals.tsv.gz")
```

### DataFrame 1 - name.basics

```
df <- read_delim(dfs[1], delim = "\t", col_names=TRUE, n_max = 5000)
df1 <- df[ , -which(names(df) == "knownForTitles")]
df1
```

### DataFrame 2 - tittle.basics

```
df2 <- read_delim(dfs[2], delim = "\t", col_names=TRUE, n_max = 5000)
df2
```

### DataFrame 3 - title.principals

```
df <- read_delim(dfs[3], delim = "\t", col_names=TRUE, n_max = 5000)
df3 <- df[ , -which(names(df) == "ordering")]
df3
```

### Variables y tipos

```
df_list <- list(df1, df2, df3)
x <- 0
for (i in df_list){
    x <- x + 1
    print(paste("DataFrame ",x))

    col_names <- colnames(i)
    col_types <- sapply(i, class)
    print(paste("Nombre de columnas:", paste(col_names, collapse = ", ")))
    print(paste("Tipo de columnas:", paste(col_types, collapse = ", ")))
}
```

## Shows en Netflix

Importe todos los registros del conjunto de datos netflix_titles.csv. Tenga en cuenta que el conjunto de datos está delimitado por",".

Asigne nombres y tipos personalizados a las variables. ¿Cuántos registros tiene el conjunto de datos?

```
df_netflix <- read_csv("/work/AiRoute/university-stuff/statistical_computing/sources/netflix_titles.csv", 
                skip = 1,
                col_names = c("Id","Tipo","Titulo", "Director", "Cast", "Pais de origen","Dia de estreno (netflix)", "Año de estreno","Rating","Duracion","Categorias","Descripcion"),
                col_types = cols(
                  Id = col_character(),
                  Tipo = col_character(),
                  Titulo = col_character(),
                  Director = col_character(),
                  Cast = col_character(),
                  `Pais de origen` = col_character(),
                  `Dia de estreno (netflix)` = col_date("%B %d, %Y"),
                  `Año de estreno` = col_integer(),
                  Rating = col_character(),
                  Duracion = col_character(),
                  Categorias = col_character(),
                  Descripcion = col_character()
                )
              )

head(df_netflix)
```

### ¿Cuántos registros tiene el conjunto de datos?

```
print(paste("Numero registros: ",nrow(df_netflix)))
```
