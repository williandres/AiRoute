```
import numpy as np
import pandas as pd
import seaborn as sns
import math
```

```
nums = np.array([38, 90.7, 70.1, 93, 60.7, 124.2, 37.8, 36.5, 45.6,
                 100.5, 91.3, 71.2, 94.1, 68.9, 90.6, 87.9, 78.9, 87.5,
                 37.8, 92.8, 72.3, 93.1, 65.3, 98.7, 90, 67.5, 98.7,
                 39, 36, 100, 95.6, 64.1, 67.5, 100.5, 98.7, 108.6,
                 41.7, 94.5, 124.9, 98.7, 54.2, 97.7, 123.9, 102.4, 117.5,
                 42.9, 95.8, 80.9, 99.5, 102.9, 38.6, 121, 101.4, 120.9,
                 43.7, 110.6, 78.8, 100.2, 123.2, 87.4, 120.8, 124.9, 99.9,
                 36.2, 57.6, 82.1, 102.4, 115.6, 125, 116.7, 50.6, 98.7,
                 76.8, 89.7, 83.4, 102.4, 112.9, 93.1, 106.8, 45.7, 39.8,
                 98.9, 120.4, 86.6, 39, 80, 70.6, 56.7, 67.8, 98.9])

print(nums)
```

```
#Creando intervalos
k = round(1 + (3.3)*(math.log10(len(nums))))
r = max(nums) - min(nums)
a = math.ceil(r / k)
p = ((((a*k) - r)/2) - min(nums)) * (-1) 
print(f'Numero de intervalos: {k} | Amplitud: {a} | Punto de partida: {p}')
```

```
intervalos = []
for i in range(0,k):
    b = p + a
    intervalos.append([p, b])
    p = b
```

```
intervalos
```

```
#label de intervalos
label = []
for v in intervalos:
    label.append(f"[{round(v[0])}, {round(v[1])})")
print(label)
```

```
#Frecuencia absoluta
fa = []
def contar(desde, hasta):
    x = 0
    for i in nums:
        if desde <= i and i < hasta:
            x += 1
    return x 

for v in intervalos:
    fa.append(contar(v[0],v[1]))
fa
```

```
#Frecuencia absoluta acumulada
faa = []
x = 0
for i in fa:
    x += i
    faa.append(x)
faa
```

```
#Frecuencia relativa
fr = np.divide(fa, len(nums))
fr
```

```
#Frecuencia relativa acumulada
fra = []
x = 0
for i in fr:
    x += i
    fra.append(x)
fra
```

```
frecuencias = pd.DataFrame({"Frecuencia absoluta" : fa,
              "Frecuencia absoluta acumulada" : faa,
              "Frecuencia relativa" : fr,
              "Frecuencia relativa acumulada" : fra
                }
            , index = label)

frecuencias
```
